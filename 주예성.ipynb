```python
maxWeight = 7
rowCount = 4

# 0은 더미데이터
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]

def knapsack():
    array = [[0 for _ in range(maxWeight + 1)] for _ in range(rowCount + 1)]

    for row in range(1, rowCount+1):
        for col in range(1, maxWeight+1):
            if weight[row] > col:  
                array[row][col] = array[row-1][col]
            else:
                value1 = money[row] + array[row-1][col-weight[row]]  
                value2 = array[row-1][col]
                array[row][col] = max(value1, value2)

    return array[rowCount][maxWeight]

knapsack()
```

```python
customer = ['정국', '뷔', '지민', '진', '슈가']

def waitingline():
    num = 0
    while True:
        num+=1
        print(f'대기 줄 상태 : {customer}')
        print(f'{customer[0]} 님 식당에 들어감')

        for i in range(0, len(customer)): # 앞 인덱스를 None으로 바꾼 후 뒤와 순차적으로 위치바꿈
            if i != 0:
                customer[i-1] = customer[i]
            customer[i] = None 

        if num == len(customer): # 끝났을 경우 줄 상태 보여주고 종료
            print(f'대기 줄 상태 : {customer}')
            print('식당 영업 종료!')
            break

waitingline()
```

```python
class TreeNode():
    def __init__ (self):
        self.left = None
        self.data = None
        self.right = None

memory = []
root = None
# dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', '요아정 요거트컵',
#             '페퍼로니피자주먹밥', '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']

sellAry = input('오늘 판매된 물건(중복O)<쉼표입력필요> --> ').split(',')

node = TreeNode()
node.data = sellAry[0]
root = node
memory.append(node)

#이진트리
for name in sellAry[1:]:
    node = TreeNode()
    node.data = name

    current = root
    while True:
        if name == current.data: break

        if name < current.data:
            if current.left == None:
                current.left = node
                memory.append(node)
                break
            current = current.left
        else:
            if current.right == None:
                current.right = node
                memory.append(node)
                break
            current = current.right

def preorder(node):
    if node == None: return
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

print(f'오늘 판매된 물건(중복O) --> {sellAry}')
print('오늘 판매된 종류(중복X) -->', end=' ')
preorder(root)
```

```python
import random

def binSearch(ary, fData):
    num = 0
    pos = -1
    start = 0
    end = len(ary) - 1

    while (start <= end):
        mid = (start + end) // 2
        
        num += 1
        print(f'## {num}회 검색 ##') # 검색횟수

        if fData == ary[mid]:
            return mid
        elif fData > ary[mid]:
            start = mid + 1
        else:
            end = mid - 1

    return pos

dataAry = []

# 100000번을 랜덤으로 0~100000 중에서 뽑음
for _ in range(100000):
    dataAry.append(random.randint(0,100000))

# 0~100000 중 랜덤으로 찾을 데이터를 뽑음
findData = random.randint(0,100000)

print(f'배열 --> {dataAry}')
position = binSearch(dataAry, findData)
if position == -1:
    print(f'{findData} (이)가 없네요.')
else:
    print(f'{findData} (은)는 {position} 위치에 있음')
```
